```python
import matplotlib.pyplot as plt
```

```python
x =[10,50,20,80]
y =[10,20,30,40] 
plt.plot(x,y)
plt.show()
```

```python
x =[10,50,20,80]
y =[10,20,30,40] 
plt.plot(x,y)
plt.xlabel('eje X')
plt.ylabel('eje Y')
plt.title('Ejemplo de grafico de lineas')
plt.show()
```

```python
x1 = [1,2,3,4]
y1 = [100,50,80,40]
plt.bar(x1,y1)
plt.show()
```

```python
x1 = ['A','B','C','D']
y1 = [100,50,80,40]
plt.bar(x1,y1)
plt.xlabel('Letras')
plt.ylabel('Valores')
plt.title('Ejemplo de grafico de Barras')
plt.show()
```

```python
x2 = ['A','B','C','D']
y2 = [100,50,80,40]
plt.pie(y2,labels=x2)
plt.title('Grafica de pie')
plt.axis('equal')#grafico circular
plt.show()
```

```python
x2 = ['A','B','C','D']
y2 = [100,50,80,40]
plt.pie(y2,labels=x2,autopct='%1.1f%%')
plt.title('Grafica de pie')
plt.axis('equal')#grafico circular
plt.show()
```

```python
x3 = [100,50,80,40]
y3 = [10,50,20,40]
z3 = [18,50,20,100]
fig = plt.figure()
graf = fig.add_subplot(111,projection='3d')
graf.plot(x3,y3,z3)
graf.set_xlabel('eje x')
graf.set_ylabel('eje y')
graf.set_zlabel('eje z')
plt.title('Emplo grafica 3d')
plt.show()
```

```python
from matplotlib import cm
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
x4 = [100,50,80,40]
y4 = [10,50,20,40]
x4,y4 = np.meshgrid(x4,y4)
z4 = np.array([
    [18,50,20,100],
    [18,50,20,100],
    [18,50,20,100],
    [18,50,20,100]
])
fig = plt.figure()
graf = fig.add_subplot(111,projection='3d')
aux = graf.plot_surface(x4,y4,z4,cmap=cm.viridis)
fig.colorbar(aux)
plt.show()
```
